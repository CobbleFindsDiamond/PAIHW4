```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory
import torch
import torch.nn as nn
import torch.nn.functional as F
import os
from sklearn.model_selection import train_test_split
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import accuracy_score, f1_score
import matplotlib.pyplot as plt
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
data = pd.read_csv("/content/seattle-weather.csv")
data.head
print(data.describe())
column_names = data.columns
print(column_names)
data['weather'] = data['weather'].map({'drizzle': 0, 'rain': 1, 'sun': 2, 'snow': 3, 'fog': 4})
data['date'] = pd.to_datetime(data['date'])
data['day_of_week'] = data['date'].dt.weekday
data['month'] = data['date'].dt.month
data['day_of_year'] = data['date'].dt.dayofyear
seq_len = 5
features = ['day_of_week', 'month', 'day_of_year', 'precipitation', 'temp_max', 'temp_min', 'wind']
X, y = [], []

for i in range(len(data) - seq_len):
    x_seq = data[features].iloc[i:i+seq_len].values
    label = data['weather'].iloc[i+seq_len]
    X.append(x_seq)
    y.append(label)

X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)

class CustRNN(nn.Module):
    def __init__(self, input_size, hidden_size):
        #calls the nn superclass which has autograd
        super(CustRNN, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        #initlizes the linear layer tunring input_size into hidden_size's size
        #it does W_x(x_t) + W_b and you get a matrix that has the dimensions of hidden_size
        #x_t is the first input
        self.Wx = nn.Linear(input_size, hidden_size)
        #so it creates initlayy random weighrs and biases to do W_x(hidden_state) + W_b whose inital hidden_state is just zeros
        self.Wh = nn.Linear(hidden_size, hidden_size)
    def forward(self, x, h_prev):
        h = torch.tanh(self.Wx(x) + self.Wh(h_prev))
        return h

class RNNModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(RNNModel, self).__init__()
        self.hidden_size = hidden_size
        self.rnn_cell = CustRNN(input_size, hidden_size)
        #fc turns the hiddent size into the correct format
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        batch_size, seq_len, input_size = x.size()
        # hidden states set to zeros
        h = torch.zeros(batch_size, self.hidden_size)

        for t in range(seq_len):
            x_t = x[:, t, :]
            h = self.rnn_cell(x_t, h)
        out = self.fc(h)
        return out
model = RNNModel(input_size=7, hidden_size=16, output_size=5)
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
num_epochs = 30
batch_size = 16
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
train_losses = []
for epoch in range(num_epochs):
    model.train()
    permutation = torch.randperm(X_train.size(0))
    total_loss = 0
    for i in range(0, X_train.size(0), batch_size):
        idx = permutation[i:i+batch_size]
        x_batch, y_batch = X_train[idx], y_train[idx]
        optimizer.zero_grad()
        outputs = model(x_batch)
        loss = loss_fn(outputs, y_batch)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    train_losses.append(total_loss)
    model.eval()
    with torch.no_grad():
        test_outputs = model(X_test)
        test_preds = torch.argmax(test_outputs, dim=1)
        acc = accuracy_score(y_test.numpy(), test_preds.numpy())
        f1 = f1_score(y_test.numpy(), test_preds.numpy(), average='weighted')
    print(f"Epoch {epoch}: Loss: {total_loss:.4f}, Accuracy: {acc:.4f}, F1 Score: {f1:.4f}")
plt.figure(figsize=(10, 6))
plt.plot(train_losses, label='RNN')
plt.title("Training Loss per Epoch (RNN)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

###Part 1
So basically have a hidden state that is passed along with the input to make a prediction. This hidden state is initially filled with zeros and then multiplied by a random weight and added with a random bias. This is added with the input which is multiplied by a random weight and added bias input into the activation function of tanh to make the next hidden state. From there pytroch handles the rest using the functions of the nn superclass to make backpropagation. To evaluate this I'm using the F1score and the Accuracy.

##Task 2 part 2
 cosine similarity lets you know how simialr directionally the 2 words are. So the higher the similarity, the more simialr they are and this can be used to make predictions since if 2 words are simialr, its also likely they are next.

```python
from transformers import BertTokenizer, BertModel
import torch
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

def bert_embedding(word):
    inputs = tokenizer(word, return_tensors="pt")
    outputs = model(**inputs)
    token_embeddings = outputs.last_hidden_state
    tokens = tokenizer.tokenize(word)
    tokenIds = tokenizer.convert_tokens_to_ids(tokens)
    output = token_embeddings[0][1:len(tokens)+1].mean(dim=0)
    return output
def batches(words):
    similarities = []
    for wordOne, wordTwo in words:
        embOne = bert_embedding(wordOne)  # (1, hidden)
        embTwo = bert_embedding(wordTwo)
        sim = torch.nn.functional.cosine_similarity(embOne, embTwo, dim=0, eps=1e-8)
        similarities.append(f"'{wordOne}', '{wordTwo}': {sim.item():.4f}")
    return similarities
def visual(words):
    embeddings = torch.stack([get_bert_embedding(w) for w in words])
    embeddings_np = embeddings.detach().numpy()
    reducer = PCA(n_components=2)
    reduced = reducer.fit_transform(embeddings_np)
    plt.figure(figsize=(8, 6))
    i = 0
    for word in words:
        x, y = reduced[i][:2]
        plt.scatter(x, y)
        plt.text(x, y, word)
        i += 1
    plt.title("visual")
    plt.xlabel("Dim 1")
    plt.ylabel("Dim 2")
    plt.grid(True)
    plt.show()

def euclideanNormDis(wordOne, wordTwo):
    return torch.norm(wordOne - wordTwo, p=2).item()
def SimOrNot(wordOne, wordTwo, similarity):
    emb1 = bert_embedding(wordOne).unsqueeze(0)
    emb2 = bert_embedding(wordTwo).unsqueeze(0)
    if similarity == True:
        return torch.nn.functional.cosine_similarity(emb1.squeeze(0), emb2.squeeze(0), dim=0).item()
    elif similarity == False:
        return euclideanNormDis(emb1, emb2)

def ranking(main_word, word_list, similarity):
    scores = []
    for word in word_list:
        score = SimOrNot(main_word, word, similarity)
        scores.append((word, score))
    return sorted(scores, key=lambda x: x[1])
def plot_word_ranking(main_word, word_list, similarity):
    ranked = ranking(main_word, word_list, similarity)
    words, scores = zip(*ranked)
    plt.figure(figsize=(10, 5))
    plt.barh(words, scores, color="skyblue")
    plt.gca().invert_yaxis()
    if similarity == True:
      way = "cos"
      other = "Similarity"
    else:
      way = "euclidean"
      other = "Dissimilarity"
    plt.title(f"'{other}' to '{main_word}' '{way}' distance)")
    plt.xlabel("Dissimilarity/Similarity Score")
    plt.tight_layout()
    plt.show()
def plot_similarity_heatmap(words, method="cosine"):
    n = len(words)
    embeddings = [get_bert_embedding(w) for w in words]
    matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            emb1, emb2 = embeddings[i], embeddings[j]
            if method == "cosine":
                sim = torch.nn.functional.cosine_similarity(emb1.unsqueeze(0), emb2.unsqueeze(0), dim=1).item()
                matrix[i][j] = sim
            if method == "euclidean":
                dist = torch.norm(emb1 - emb2, p=2).item()
                matrix[i][j] = dist
    plot_word_ranking("dog", words, True)
    plot_word_ranking("dog", words, False)
    plt.figure(figsize=(8, 6))
    ax = sns.heatmap(matrix, annot=True, xticklabels=words, yticklabels=words, cmap="viridis")
    title = f"{method.capitalize()} {'Similarity' if method == 'cosine' else 'Dissimilarity'} Heatmap"
    plt.title(title)
    plt.show()
words = ["king", "thing", "man", "woman", "nuts", "straw", "car", "ships", "doctor"]
plot_similarity_heatmap(words, method="cosine")      # for cosine similarity
plot_similarity_heatmap(words, method="euclidean")   # for euclidean dissimilarity
```
